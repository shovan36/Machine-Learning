```python
import matplotlib.pyplot as plt
```

```python
y = titanic[titanic['Survived']==1].groupby('Sex')['Name'].count()
x=['Females', 'Males']
```

```python
print(y)
```

```python
plt.bar(x,y) 
plt.title('survival')
```

```python
y = titanic[titanic['Survived']==1].groupby('Pclass')['Name'].count()
x=['Class 1', 'Class 2', 'Class 3']
plt.bar(x,y)
```

```python
childs = titanic[(titanic['Survived']==1) & (titanic['Age']<12)]['Name'].count()
adults = titanic[(titanic['Survived']==1) & (titanic['Age']>12)]['Name'].count()
```

```python
y=[childs,adults]
x=['childs','adults']
plt.bar(x,y)
```
