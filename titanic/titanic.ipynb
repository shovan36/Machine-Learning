```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.tail()
```

```python
titanic['Age'].isnull().sum()
```

```python
titanic['Age'].fillna(titanic['Age'].median(),inplace= True)
```

```python
titanic.tail()
```

```python
plt.hist(titanic.Age)
```

```python
after_removal = titanic[titanic['Age']<100]
```

```python
titanic.__len__()
```

```python
after_removal.__len__()
```

```python
titanic.head()
```

```python
titanic['Sex'][titanic['Sex']=='male']=1
titanic['Sex'][titanic['Sex']=='female']=0

```

```python
titanic.head()
```

```python
embaraked = pd.get_dummies(titanic['Embarked'],prefix = 'Embarked')
```

```python
new_titanic = pd.concat([titanic,embaraked],axis=1)
```

```python
new_titanic.head()
```

```python
new_titanic.describe()
```

```python
new_titanic.drop(['PassengerId','Name','Ticket','Embarked','Cabin'],axis=1,inplace=True)
```

```python
new_titanic.head()
```

```python
new_titanic.to_csv('final_dataset.csv',index= False)
```
