```python
import tensorflow as tf
from tensorflow.keras import layers
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
titanic =pd.read_csv('final_dataset.csv')
dataset=titanic.copy()
X=np.asarray(dataset.drop('Survived',1))
y=np.asarray(dataset['Survived'])
```

```python


#splliting data for training and testing
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 0)
```

```python
#normalizing the data by keeping them in standard scale
sc= StandardScaler()
X_train_normalized = sc.fit_transform(X_train)
X_test_normalized = sc.transform(X_test)
```

```python
#defining the neural network architechture

numDims = X_train_normalized.shape[1] #gets the number of features of the data
#fully conected architecture hence use dense
dnnModel = tf.keras.Sequential([
    layers.Dense(15,activation='elu',input_shape =(numDims,)),
    
    layers.Dropout(0.3),
    layers.Dense(19,activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.21),
    layers.Dense(120,activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(160,activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(20,activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.1),
    layers.BatchNormalization(),
    layers.Dense(2) #binary classifiacation problem hence only 2 neurons in output layer
    
])
optimizer = tf.keras.optimizers.RMSprop(0.01)
dnnModel.compile(optimizer=optimizer,loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits= True),metrics = ['accuracy'])
```

```python
dnnModel.summary()
```

```python
dnnModel.fit(X_train_normalized , y_train, epochs =100)
```

```python
#checking test accuracy
test_loss,test_acc = dnnModel.evaluate(X_test_normalized , y_test, verbose=2)
print('\nTest accuracy :', test_acc)
```

```python
y_pred = dnnModel.predict_classes(X_test)
```
